```python
mixer.init()
sound= mixer.Sound(r"C:\Users\91989\project expo\MINI_PROJECT_2023\mrlEyes_2018_01\emergency-alarm-with-reverb-29431.mp3")
cap = cv2.VideoCapture(r"C:\Users\91989\project expo\MINI_PROJECT_2023\mrlEyes_2018_01\test123.mp4")
Score = 0

while True:
    ret, frame = cap.read()
    height, width = frame.shape[0:2]
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=3)
    eyes = eye_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=1)

    #cv2.rectangle(frame, (0,height-50),(200,height),(0,0,0),thickness=cv2.FILLED)
    
    for (a, b, c, e) in faces:
        cv2.rectangle(frame, pt1=(a, b), pt2=(a + c, b + e), color=(255, 0, 0), thickness=3)
        
    for (ia, ib, ic, ie) in eyes:
        cv2.rectangle(frame, pt1=(ia, ib), pt2=(ia + ic, ib + ie), color=(0, 255, 0), thickness=3)
        
        # Preprocessing step
        eye = frame[ib:ib+ie, ia:ia+ic]
        eye = cv2.resize(eye, (80, 80))
        eye = eye / 255.0
        eye = eye.reshape(1, 80, 80, 3)
        
        
        # Model prediction
        # Replace 'model' with your actual model
        prediction = model.predict(eye)
        
         # if eyes are closed
        if prediction[0][0]>0.20:
            cv2.putText(frame,'closed',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            cv2.putText(frame,'Score'+str(Score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            Score=Score+1
            if(Score>15):
                try:
                    sound.play()
                except:
                    pass
            
        # if eyes are open
        elif prediction[0][1]>0.90:
            cv2.putText(frame,'open',(10,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)      
            cv2.putText(frame,'Score'+str(Score),(100,height-20),fontFace=cv2.FONT_HERSHEY_COMPLEX_SMALL,fontScale=1,color=(255,255,255),
                       thickness=1,lineType=cv2.LINE_AA)
            Score = Score-1
            if (Score<0):
                Score=0
        
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
